```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

from glob import glob
from tqdm import tqdm
```

```python
fps = glob('../slides/3_sam_masks_big_cell/*')
for fp in tqdm(fps):
    
    wing_name = fp.split('/')[-1].split('.')[0]
    mask = cv2.imread(fp, cv2.IMREAD_GRAYSCALE)
    contours, hierarchy = cv2.findContours((mask > 0).astype('uint8'), 1, 2)
    try:
        biggest_contour = contours[np.argmax([cv2.contourArea(x) for x in contours])]
        
        mask = np.zeros(mask.shape)
        cv2.drawContours(mask, [biggest_contour], 0, 1, -1)
        #plt.imshow((mask>0))
        new_fp = '../slides/4_cleaned_sam_masks_big_cell/' + wing_name + '.png'
        cv2.imwrite(new_fp, mask*255)
    except Exception as ex:
        print(ex)
        pass

```
