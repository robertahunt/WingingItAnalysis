```python
"""
This notebook visualizes the age predictions from the deep learning model 
"""
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../age_pred/logs/WingingItResNet/version_3/Epoch_99_pred.csv', index_col=0)
train = df[df['set'] == 'train']
val = df[df['set'] == 'val']
plt.scatter(train['labels'], train['preds'], alpha=0.5, label='train')
plt.xlabel('Age')
plt.ylabel('Prediction')

plt.scatter(val['labels'], val['preds'], label='val', color='r', alpha=0.5)
plt.legend()
```
