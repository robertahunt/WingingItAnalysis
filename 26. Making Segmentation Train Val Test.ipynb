```python
import os
import cv2
import json
import shutil
import numpy as np
import matplotlib.pyplot as plt


from tqdm import tqdm
from glob import glob
from sklearn.model_selection import train_test_split
```

```python
name_to_class = {'FWL':1, 'MC':2, '1sMC':3, '2sMC':4,'3sMC':5,'2MdC':6}
cells = name_to_class.keys()

for fn in tqdm(good_images):
    crop_fp = '../2_live_bees/6_cropped_and_flipped/' + fn
    crop = cv2.imread(crop_fp)

    mask = np.zeros(crop.shape[:2]).astype('uint8')
    
    for cell in cells:
        cell_mask_fp = f'../2_live_bees/7_{cell}_masks/' + fn
        cell_mask = cv2.imread(cell_mask_fp, cv2.IMREAD_GRAYSCALE)
        mask[np.where(cell_mask > 0.5)] = name_to_class[cell]

    meta_fp = '../2_live_bees/3_card_mask_matches_metadata/' + fn.replace('.png','.json')
    with open(meta_fp, 'r') as f:
        meta = json.load(f)

    if not meta['is_facing_up?']:
        mask = np.flipud(mask)

    
    cv2.imwrite('../2_live_bees/8_check_masks/' + fn,mask)
```

```python
fps = glob('../2_live_bees/8_check_masks_manual_corrections/*')
fns = [x.split('/')[-1] for x in fps]
train, val = train_test_split(fns, random_state=42, train_size=0.9)
for fps, _set in zip([train, val],['train','val']):
    for fn in fps:
        fp = '../2_live_bees/6_cropped_and_flipped/' + fn
        mask_fp = '../2_live_bees/8_check_masks_manual_corrections/' + fn
        if not os.path.exists(f'../2_live_bees/8_train_val_split/{_set}/images/'):
            os.makedirs(f'../2_live_bees/8_train_val_split/{_set}/images/')
        if not os.path.exists(f'../2_live_bees/8_train_val_split/{_set}/masks/'):
            os.makedirs(f'../2_live_bees/8_train_val_split/{_set}/masks/')
        
        new_fp = f'../2_live_bees/8_train_val_split/{_set}/images/' + fn
        new_mask_fp = f'../2_live_bees/8_train_val_split/{_set}/masks/' + fn

        shutil.copyfile(fp, new_fp)
        shutil.copyfile(mask_fp, new_mask_fp)
```
